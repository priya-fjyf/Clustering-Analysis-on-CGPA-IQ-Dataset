## 1. Import Necessary Libraries

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn
```

## 2. Load Dataset

```python
df = pd.read_csv("/kaggle/input/cgpa-and-iq-clustering-dataset/CGPA_IQ_Clustering_Dataset_200.csv")
```

```python
df
```

## 3. Distribution of Points of Dataset

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='CGPA', y='IQ', color='blue', s=50)

# Add labels and title
plt.title('Scatter Plot of CGPA vs IQ')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.grid(True)
plt.show()

```

## 4. Import Algorithm from Sklearn

```python
from sklearn.cluster import KMeans
```

## 5. Calculate Wcss using for loop and decide How many number of Clusters is best for Dataset

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

## 6. To Check how many numbers of Culsters we can used

```python
plt.plot(range(1,11),wcss)
```

## 7. When you decide the graph is stablise you can choose 5 or 6 as a no. of clusters

```python
X = df.iloc[:,:].values
kmeans = KMeans(n_clusters=5)
y_means = kmeans.fit_predict(X) 
```

```python
y_means
```

## 8. Using Scatter to plot how the Clusters is done  

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],color = 'blue')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],color = 'red')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],color = 'green')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],color = 'yellow')
plt.scatter(X[y_means == 4,0],X[y_means == 4,1],color = 'black')



plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.grid(True)
plt.show()
```

## 9. To check accuracy i use sillhouette_score

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(X, y_means)
print("Silhouette Score:", score)
```
